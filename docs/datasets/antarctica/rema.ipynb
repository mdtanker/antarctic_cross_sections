# REMA surface elevations

**Description:**  Reference Elevation Model of Antarctica, referenced to the WGS84 ellipsoid.

**Dataset DOI:** N.A.

**Associated paper:** The Reference Elevation Model of Antarctica. DOI: 10.5194/tc-13-665-2019

```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, utils
```

```python
version_names = [
    "500m",
    "1km",
]

grids = []
for name in version_names:
    data = fetch.rema(
        version=name,
        # available options
        # region,
        # spacing,
        # registration ("g" for gridline or "p" for pixel),
    )
    grids.append(data)
```

```python
_ = utils.get_grid_info(data, print_info=True)
```

```python
fig = maps.plot_grd(
    grids[0],
    title="REMA (500m resolution)",
    coast=True,
    cbar_label="Surface elevation (meters)",
    hist=True,
)
fig = maps.plot_grd(
    grids[1],
    fig=fig,
    origin_shift="xshift",
    title="REMA (1km resolution)",
    coast=True,
    cbar_label="Surface elevation (meters)",
    hist=True,
)
fig.show()
```
