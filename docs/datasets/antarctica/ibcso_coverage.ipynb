```python
%%capture
%load_ext autoreload
%autoreload 2


from polartoolkit import fetch, maps, regions
```

```python
points_gdf, polygons_gdf = fetch.ibcso_coverage(
    # available options
    region=regions.antarctica,
)
```

```python
points_gdf.head()
```

```python
polygons_gdf.head()
```

```python
fig = maps.basemap(
    region=regions.antarctica,
    coast=True,
    coast_pen="0.1,black",
    title="IBCSO points",
    points=points_gdf,
    points_style="p",
    points_fill="cyan",
    hemisphere="south",
)
fig.show(dpi=200)
```
