**Associated papers:** 

S-Velocity Model and Inferred Moho Topography beneath the Antarctic Plate from Rayleigh Waves. DOI: 10.1002/2014JB011332

```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, regions, utils
```

```python
version_names = [
    "an-2015",
    # "shen-2018",
]

grids = []
for name in version_names:
    data = fetch.crustal_thickness(
        version=name,
        # available options
        # region,
        # spacing,
        # registration ("g" for gridline or "p" for pixel),
    )
    grids.append(data)
```

```python
_ = utils.get_grid_info(data, print_info=True)
```

```python
fig = maps.plot_grd(
    data,
    title="Crustal Thickness (An et al. 2015)",
    region=regions.antarctica,
    cbar_label="meters",
    coast=True,
    hist=True,
)

fig.show()
```
