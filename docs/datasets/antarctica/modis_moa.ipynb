# MODIS MOA 

**Description:** MODIS optical image mosaic of Antarctica.

Requires an EarthData login, see [link](../../how_to/download_NASA_data.ipynb) for how to configure this. 

**Dataset:** 

Haran, T., J. Bohlander, T A Scambos, T. Painter, and M. Fahnestock. “MODIS Mosaic of Antarctica 2008-2009 (MOA2009) Image Map, Version 2,” 2021. https://doi.org/10.5067/4ZL43A4619AF.

**Associated paper:** 

Scambos, T.A., T.M. Haran, M.A. Fahnestock, T.H. Painter, and J. Bohlander. “MODIS-Based Mosaic of Antarctica (MOA) Data Sets: Continent-Wide Surface Morphology and Snow Grain Size.” Remote Sensing of Environment 111, no. 2–3 (2007): 242–57. https://doi.org/10.1016/j.rse.2006.12.020.


```python
%%capture
%load_ext autoreload
%autoreload 2
import os

from polartoolkit import fetch, maps, utils
```

```python
# set default to southern hemisphere for this notebook
os.environ["POLARTOOLKIT_HEMISPHERE"] = "south"
```

```python
versions = ["750m", "125m"]


grids = []
for name in versions:
    data = fetch.modis(
        version=name,
    )
    grids.append(data)
    print(f"Info for {name}")
    _ = utils.get_grid_info(data, print_info=True)
    print("##########")
```

```python
fig = maps.subplots(
    grids,
    fig_title="MODIS Mosaic of Antarctica",
    subplot_titles=versions,
    modis=True,  # this automatically uses a nice colorscale for modis data
    coast=True,
    coast_pen="0.6p,salmon",
)

fig.show(dpi=200)
```
