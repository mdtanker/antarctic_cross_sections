```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, regions, utils
```

```python
version_names = [
    "shen-2018",
    "an-2015",
]

grids = []
for name in version_names:
    data = fetch.moho(
        version=name,
        # available options
        # region,
        # spacing,
        # registration ("g" for gridline or "p" for pixel),
    )
    grids.append(data)
    print(f"Info for {name}")
    _ = utils.get_grid_info(data, print_info=True)
    print("##########")
```

```python
cpt_lims = utils.get_combined_min_max(grids, robust=True)

fig = maps.plot_grd(
    grids[0],
    title="Moho Depth (Shen et al. 2018)",
    region=regions.antarctica,
    cmap="acton",
    cbar_label="meters",
    coast=True,
    hist=True,
    cpt_lims=cpt_lims,
    hemisphere="south",
)
fig = maps.plot_grd(
    grids[1],
    fig=fig,
    origin_shift="xshift",
    title="Moho Depth (An et al. 2015)",
    region=regions.antarctica,
    cmap="acton",
    cbar_label="meters",
    coast=True,
    hist=True,
    cpt_lims=cpt_lims,
    hemisphere="south",
)
fig.show()
```
