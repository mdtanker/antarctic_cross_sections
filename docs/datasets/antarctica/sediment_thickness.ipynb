# Sediment Thickness
**Description:**  Sediment thickness grids from various sources.

**Dataset DOIs:** 
Tankersley et al. 2022: 10.1594/PANGAEA.941238

Lindeque et al. 2016: 10.1594/PANGAEA.864906

**Associated papers:** 
ANTASed -- An Updated Sediment Model for Antarctica. DOI: 10.3389/feart.2021.722699

Basement Topography and Sediment Thickness Beneath Antarctica's Ross Ice Shelf. DOI: 10.1029/2021GL097371

Preglacial to glacial sediment thickness grids for the Southern Pacific Margin of West Antarctica. DOI:10.1002/2016GC006401

GlobSed: Updated Total Sediment Thickness in the World's Oceans. DOI: 10.1029/2018GC008115

```python
%%capture
%load_ext autoreload
%autoreload 2


from polartoolkit import fetch, maps, regions, utils
```

```python
version_names = [
    "ANTASed",
    "tankersley-2022",
    "lindeque-2016",
    "GlobSed",
]

grids = []
for name in version_names:
    data = fetch.sediment_thickness(
        version=name,
        # available options
        # region,
        # spacing,
        # registration ("g" for gridline or "p" for pixel),
    )
    grids.append(data)
    print(f"Info for {name}")
    _ = utils.get_grid_info(data, print_info=True)
    print("##########")
```

```python
fig = maps.subplots(
    grids,
    region=regions.antarctica,
    fig_title="Sediment Thickness Models",
    subplot_titles=version_names,
    cbar_labels=["meters" for x in version_names],
    margins=".5c/1.3c",
    coast=True,
)
fig.show()
```
