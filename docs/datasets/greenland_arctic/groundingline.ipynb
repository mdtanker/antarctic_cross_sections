```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, regions
```

```python
version_names = [
    "BAS",
    "measures-greenland",
]

filepaths = []
for name in version_names:
    filepaths.append(
        fetch.groundingline(
            version=name,
        )
    )
```

```python
fig = maps.basemap(
    region=regions.greenland,
    hemisphere="north",
)

# plot the shapefiles
pens = [".8p,red", ".4p,blue"]
for i, v in enumerate(version_names):
    fig.plot(filepaths[i], pen=pens[i], label=v)

# add a legend
fig.legend()

fig.show(dpi=200)
```
