```python
%%capture
%load_ext autoreload
%autoreload 2

import os

from polartoolkit import fetch, maps, utils
```

```python
# set default to northern hemisphere for this notebook
os.environ["POLARTOOLKIT_HEMISPHERE"] = "north"
```

```python
versions = ["500m", "100m"]

grids = []
for name in versions:
    data = fetch.modis(
        version=name,
    )
    grids.append(data)
    print(f"Info for {name}")
    _ = utils.get_grid_info(data, print_info=True)
    print("##########")
```

```python
fig = maps.subplots(
    grids,
    fig_title="MODIS Mosaic of Greenland",
    subplot_titles=versions,
    modis=True,  # this automatically uses a nice colorscale for modis data
    coast=True,
    coast_pen="0.6p,salmon",
)

fig.show(dpi=200)
```
