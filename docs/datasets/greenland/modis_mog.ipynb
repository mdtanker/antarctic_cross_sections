# MODIS MOG

**Description:** MODIS optical image mosaic of Greenland.

**Dataset DOI:** 10.5067/9ZO79PHOTYE5

**Associated paper:** N.A.

```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, regions
```

```python
versions = ["500m", "100m"]


grids = []
for name in versions:
    data = fetch.modis_mog(
        version=name,
    )
    grids.append(data)
```

```python
fig = maps.subplots(
    grids,
    region=regions.greenland,
    fig_title="MODIS Mosaic of Greenland",
    subplot_titles=versions,
    cmaps=["grayC", "grayC"],
    image=True,
)

fig.show()
```
