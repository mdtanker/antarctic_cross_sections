```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import fetch, maps, utils, regions
```

```python
version_names = [
    "BAS",
    "measures-greenland",
]

filepaths = []
for name in version_names:
    filepaths.append(fetch.groundingline(
        version=name,
    ))
```

```python
fig = maps.basemap(
    region=regions.greenland,
    hemisphere="south",
)

pens = [".8p,red", ".4p,blue"]
# plot the shapefiles
for i, v in enumerate(version_names):
    fig.plot(filepaths[i], pen=pens[i], label=v)


fig.show()
```
