# Add map to profile plot

```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import profile
```

```python
# define profile endpoints in EPSG:3413 meters
a = (307e3, -1960e3)
b = (802e3, -2000e3)

fig, _, _ = profile.plot_profile(
    method="points",
    start=a,
    stop=b,
    num=1000,
    add_map=True,
    layers_legend_scale=1,
    hemisphere="north",
)
fig.show()
```
