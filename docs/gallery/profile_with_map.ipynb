# Add map to profile plot

```python
%%capture
%load_ext autoreload
%autoreload 2

from polartoolkit import profile
```

```python
# define profile endpoints in EPSG:3031 meters
a = (-590e3, -1070e3)
b = (-100e3, -545e3)

fig, _, _ = profile.plot_profile(
    method="points",
    start=a,
    stop=b,
    num=100,
    add_map=True,
    layers_legend_scale=1,
)
fig.show()
```
